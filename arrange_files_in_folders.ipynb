```python
import os
from pathlib import Path
import glob
import shutil
```

```python
root_folder = os.path.join(os.path.abspath(os.curdir),'data_set')
main_folder = os.path.join(root_folder,'train')

files = os.listdir(main_folder)

folders_names = list(set([f.split("_")[0] for f in files]))
for folder_name in folders_names:
    abs_folder_path = os.path.join(main_folder,folder_name)
    Path(abs_folder_path).mkdir(parents=False, exist_ok=True)
    files2move = glob.glob(os.path.join(main_folder,folder_name+"_*.wrl"))
    for file in files2move:
        dst = os.path.join(abs_folder_path,file.split(os.sep)[-1])
        src = file
        # print(dst,src)
        shutil.move(src,dst)
    print(len(files2move))

    
#Path(models_folder).mkdir(parents=True, exist_ok=True)
```
